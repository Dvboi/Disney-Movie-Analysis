```python

# We can see that wherever Production company is null we have values in Production companies, 
# which is intuitive enough by it's name itself, so we can merge it into one column.
# Same thing was found for country & countries , Language & Languages.

data['Production company'] = np.where(data['Production company'].isnull(),
                                     data['Production companies'],data['Production company']) 

data['Country'] = np.where(data['Country'].isnull(),data['Countries'],data['Country']) 

data['Language'] = np.where(data['Language'].isnull(),data['Languages'],data['Language']) 

# Also narrated by has a lot of nulls and box_office_api is unreliable being a third party API data,
# whereas wiki is far more trustworthy. 

data.drop(['Production companies','Languages','Countries','Narrated by','BoxOffice_Api']
           ,axis=1,inplace=True)
```

Cleaning Columns from here

```python
# remove [1]/[2] link texts and brackets
data['Production company'] = data['Production company'].astype("string").apply(lambda x:re.sub(r'(\[\d\])*|(\[|\])*','',str(x)))             
```

```python
# Cleaning dates
data['Release date'] = data['Release date'].apply(lambda x:re.sub(r'(\[\d\])*','',str(x))) # remove [1]/[2] link texts
def catch_date(x):
    pattern1 = r'\b(\d{4}-\d{2}-\d{2})'  # to catch 1960-04-13
    pattern2 = r'\b([a-zA-Z]+ \d{1,2}, \d{4})'  # to catch 'February 02, 1980'
    if len(re.findall(pattern1,str(x)))!=0:
        return re.findall(pattern1,str(x))[0]
    elif len(re.findall(pattern2,str(x)))!=0:
        return re.findall(pattern2,str(x))[0]
    else:
        return np.nan

data['Release date'] = data['Release date'].astype("string").apply(catch_date)               
```

```python
pattern_rt = r'(\d{2,3} minutes|\d{2,3} min|\d{2,3} min\.)'  # parse out minutes
data['Running time'] = data['Running time'].apply(lambda x:re.findall(pattern_rt,str(x))[0] if len(re.findall(pattern_rt,str(x)))!=0 else np.nan)                                  
```

```python

# converting values   United StatesUnited Kingdom to --------> ['United States','United Kingdom'] 

def Country_cleaner(x):
    lst = ['United States','United Kingdom','Mexico','Spain', 'Italy','France','Australia','India',
      'Germany','Russia','China','Canada','Malaysia','Japan','Norway','Sweden']
    if type(x)==np.float:  # to handle NAN's
        return x
    temp = []
    for ele in lst:
        if len(re.findall(r'%s'%(ele),x))>0:
            temp.append(ele)
    return temp
```

```python
data.Country = data.Country.apply(Country_cleaner)
```

```python

# converting values   EnglishSpanish to --------> ['English','Spanish'] 


def Language_cleaner(x):
    lst = ['English', 'Spanish', 'Portuguese','Norwegian','Inuktitut','Japanese',
      'French',  'Mandarin' ,'Chinese','Hindi','Luganda','German','Swedish' , 'Danish' ,'Icelandic',
      'Gaelic','Scots']
    if type(x)==np.float:
        return x
    temp = []
    for ele in lst:
        if len(re.findall(r'%s'%(ele),x))>0:
            temp.append(ele)
    return temp
```

```python
data.Language = data.Language.apply(Language_cleaner)
```

```python
pattern2 = r'\$\d+,?\d+,?\d+,?\d+\.?'  # for $40,070,995, $950000
pattern3 = r'\$\d+\.?\d*\s?(?:million|billion|thousand)?' # for $178 million[3]
pattern1 = r'\$\d+\.?\d*–\$?\d+\.?\d* million'  # for ranges of money $76.4–$83.3 million[4][5] , later took the average after parsing out the values
pattern4 = r'₹?\s?\d+ crore|lakh|lakhs'   # to grab indian rupee format
pattern5 = r'₹\d+\.?\d*? (?:billion|million)'  # to grab ₹1.09 billion
final_regex = '(%s|%s|%s|%s|%s)' %(pattern1,pattern2,pattern3,pattern4,pattern5)

def clean_money(ele):
    if type(ele)==np.float:
        return ele
    return re.findall(final_regex,ele)
        
```

```python
data['Box office'] = data['Box office'].apply(clean_money)
data['Box office'] = data['Box office'].apply(lambda x:np.nan if type(x)==np.float else x[-1])
```

```python
# same parsing for budget
data['Budget'] = data['Budget'].apply(lambda x: np.nan if type(x)==np.float else re.sub(r'(\[\d\])*','',str(x))) # remove [1]/[2] link texts
data['Budget'] = data['Budget'].apply(clean_money) 
data.loc[158,'Budget'] = np.nan  # edge case where budget value was ['Unknown']
data['Budget'] = data['Budget'].apply(lambda x:np.nan if type(x)==np.float else x[-1])
```

```python
# cleaning and converting to numeric
data['Rotten_Tomatoes(in %)'] = data.Rotten_Tomatoes.apply(lambda x:np.nan if type(x)==np.float else int(str(x).replace('%','')))
data.drop('Rotten_Tomatoes',axis=1,inplace=True)
```

```python
# all these columns require same treatment
temp = ['Based on', 'Starring', 'Music by',
       'Distributed by', 'Story by', 'Cinematography', 'Edited by',
       'Screenplay by','Directed by','Produced by','Written by']
for col in temp:
    data[col] = data[col].apply(lambda x: np.nan if type(x)==np.float else re.sub(r'(\[\d\])*','',str(x))) # remove [1]/[2] link texts
    data[col] = data[col].apply(lambda ele:np.nan if type(ele)==np.float else ele[2:-2].replace('\'',''))
```

```python
# converting currency
def converter(ele):
    if type(ele)==np.float:
        return np.nan
    if '$' in ele:
        if '–' in ele:
            t = re.sub(r'[^\d+\.?\d+–]','',ele)
            temp = t.split('–')
            temp = (float(temp[0]) + float(temp[1])) / 2
            return temp*1000000
        if '-' in ele:
            t = re.sub(r'[^\d+\.?\d+-]','',ele)
            temp = t.split('-')
            temp = (float(temp[0]) + float(temp[1])) / 2
            return temp*1000000
        if 'million' in ele:
            t = re.sub(r'[^\d+\.?\d+]','',ele)
            return float(t)*1000000
        if 'billion' in ele:
            t = re.sub(r'[^\d+\.?\d+]','',ele)
            return float(t)*1000000000
        if ',' in ele:
            t = ele.replace('$','').replace(',','')
            return int(t)
        else:
            t = float(ele.replace('$','')) * 1000000
            return t
    if '₹' in ele:
        if 'billion' in ele:
            t = re.sub(r'[^\d+\.?\d+]','',ele)
            return float(t)*13722010
        if 'million' in ele:
            t = re.sub(r'[^\d+\.?\d+]','',ele)
            return float(t)*13616.61
        if 'crore' in ele:
            t = re.sub(r'[^\d+\.?\d+]','',ele)
            return float(t)*136300.78
    else:   # to catch 50 crore
        t = re.sub(r'[^\d+\.?\d+]','',ele) 
        return float(t)*136300.78
```

```python
data['Budget'] = data['Budget'].apply(converter)
data['Box office'] = data['Box office'].apply(converter)
```

```python
# helper fuction for cleaning and converting to numeric
def run_time(x):
    if type(x)==np.float:
        return x
    t = re.sub('[^\d+]','',x)
    return int(t)
```

```python
data['Running time'] = data['Running time'].apply(run_time)
```
